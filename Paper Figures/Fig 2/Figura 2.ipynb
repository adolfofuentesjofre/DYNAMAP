```python
import svgutils.transform as st
from svgutils.compose import Text
import cairosvg

# Layout settings
TARGET_WIDTH = 400
TARGET_HEIGHT = 400
LEFT_MARGIN   = 30
RIGHT_MARGIN  = 30
TOP_MARGIN    = 30
BOTTOM_MARGIN = 30
TITLE_MARGIN  = 20  # space reserved above each plot for the label

files = [
    ("Output/2-Fig2-A-NOMINATENN-Figure.svg",          (0,   0),   "a)"),
    ("Output/3-Fig2-B-Learning-Figure.svg",            (300, 0),   "b)"),
    ("Output/4-Fig2-C-People-Proposal-Figure.svg",     (0,   300), "c)"),
    ("Output/5-Fig2-D-Boxplot-Figure.svg",             (300, 300), "d)", (395, 395)),
    ("Output/6-Fig2-E-Feauture-Importance-Figure.svg", (300, 450), "e)", (395, 395)),
]

def safe_get_size(fig, default_w=800, default_h=600):
    try:
        w, h = fig.get_size()
        w = float(w.replace("px", "")) if w else default_w
        h = float(h.replace("px", "")) if h else default_h
        return w, h
    except Exception:
        return default_w, default_h

def scale_to_fit(element, orig_w, orig_h, target_w, target_h):
    scale = min(target_w / orig_w, target_h / orig_h)
    element.scale(scale)
    return orig_w * scale, orig_h * scale  # scaled_w, scaled_h

elements = []
max_x = 0
max_y = 0

for file_info in files:
    if len(file_info) == 4:
        filepath, (x, y), label, (custom_width, custom_height) = file_info
    else:
        filepath, (x, y), label = file_info
        custom_width, custom_height = TARGET_WIDTH, TARGET_HEIGHT

    # shift by margins
    x_pos = x + LEFT_MARGIN
    y_pos = y + TOP_MARGIN

    fig = st.fromfile(filepath)
    root = fig.getroot()
    root.root.attrib.pop("transform", None)

    w, h = safe_get_size(fig)
    scaled_w, scaled_h = scale_to_fit(root, w, h, custom_width, custom_height)

    root.moveto(x_pos, y_pos + TITLE_MARGIN)
    label_text = Text(label, x_pos + 5, y_pos + 15, size=16, weight="bold")

    elements.append(root)
    elements.append(label_text)

    max_x = max(max_x, x_pos + scaled_w)
    max_y = max(max_y, y_pos + TITLE_MARGIN + scaled_h)

final_width  = int(max_x + RIGHT_MARGIN)
final_height = int(max_y + BOTTOM_MARGIN)

combined_fig = st.SVGFigure(f"{final_width}px", f"{final_height}px")
combined_fig.root.attrib["viewBox"] = f"0 0 {final_width} {final_height}"
combined_fig.append(elements)

svg_path = "Output/combined_figure.svg"
pdf_path = "Output/combined_figure.pdf"

combined_fig.save(svg_path)
cairosvg.svg2pdf(url=svg_path, write_to=pdf_path)

print(f"✅ Combined PDF saved successfully: {pdf_path} (canvas: {final_width}×{final_height})")
```
