```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
user_10 = pd.read_excel('rendimiento_10_usuarios.xlsx')
user_25 = pd.read_excel('rendimiento_25_usuarios.xlsx')
user_50 = pd.read_excel('rendimiento_50_usuarios.xlsx')
user_75 = pd.read_excel('rendimiento_75_usuarios.xlsx')
user_100 = pd.read_excel('rendimiento_100_usuarios.xlsx')
user_250 = pd.read_excel('rendimiento_250_usuarios.xlsx')
user_500 = pd.read_excel('rendimiento_500_usuarios.xlsx')
user_1000 = pd.read_excel('rendimiento_1000_usuarios.xlsx')
```

```python

# Preparar los datos para la gráfica
data_acc = pd.DataFrame({
    'epochs': user_10['epochs'],
    '10 users': user_10['acc train'],
    '50 users': user_50['acc train'],
    '100 users': user_100['acc train'],
    '250 users': user_250['acc train'],
    '500 users': user_500['acc train'],
    '1000 users': user_1000['acc train']
})

data_loss = pd.DataFrame({
    'epochs': user_10['epochs'],
    '10 users': user_10['loss train'],
    '50 users': user_50['loss train'],
    '100 users': user_100['loss train'],
    '250 users': user_250['loss train'],
    '500 users': user_500['loss train'],
    '1000 users': user_1000['loss train']
})


# Crear las figuras con seaborn
sns.set(style="whitegrid")

# Figura 1: accuracy
plt.figure(figsize=(12, 8))
sns.lineplot(x='epochs', y='value', hue='variable', style='variable', markers=True, dashes=False, 
             data=pd.melt(data_acc, ['epochs']))
plt.title('Training Accuracy Over Epochs',fontsize=20)
plt.xlabel('Epochs',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.legend(title='Train',fontsize=12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()

# Figura 2: loss
plt.figure(figsize=(12, 8))
sns.lineplot(x='epochs', y='value', hue='variable', style='variable', markers=True, dashes=False, 
             data=pd.melt(data_loss, ['epochs']))
plt.title('Training Loss Over Epochs',fontsize=20)
plt.xlabel('Epochs',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.legend(title='Train',fontsize=12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python

# Cargar y preprocesar los datos de coordenadas para cada par de archivos
def preprocess_coords(coordsP_file, coordsU_file):
    # Cargar los datos de coordenadas
    coordsP = pd.read_csv(coordsP_file)
    coordsU = pd.read_csv(coordsU_file)

    # Filtrar datos extremos en coordsU
    ext_iz = coordsU[(coordsU['politica'] >= 0) & (coordsU['politica'] <= 1)]
    ext_der = coordsU[(coordsU['politica'] >= 9) & (coordsU['politica'] <= 10)]
    ext_iz['Political parties'] = 'Far-left [0,1]'
    ext_der['Political parties'] = 'Far-right [9,10]'
    extremos = pd.concat([ext_iz, ext_der], axis=0)

    return coordsP, extremos

# Definir la lista de archivos para cada par de coordenadas
files_list = [
    ('weights/coordsP_epoch_0_begin.csv', 'weights/coordsU_epoch_0_begin.csv'),
    ('weights/coordsP_epoch_5.csv', 'weights/coordsU_epoch_5.csv'),
    ('weights/coordsP_epoch_10.csv', 'weights/coordsU_epoch_10.csv'),
#    ('weights/coordsP_epoch_15.csv', 'weights/coordsU_epoch_15.csv'),
    ('weights/coordsP_epoch_20.csv', 'weights/coordsU_epoch_20.csv')
]


epochs = [0,5,10,20]
# Iterar sobre los archivos, preprocesar y plotear
for idx, (coordsP_file, coordsU_file) in enumerate(files_list):
    plt.figure(idx)
    # Preprocesar los datos de coordenadas
    coordsP, extremos = preprocess_coords(coordsP_file, coordsU_file)

    # Plot para coordsU
    sns.kdeplot(data=extremos, x='z2', y='z1', hue='Political parties', palette=["red", "blue"])
    sns.scatterplot(data=extremos, x='z2', y='z1', alpha=0.3, hue='Political parties', palette=["red", "blue"])
    plt.xlabel('Political Dimension 1',fontsize=16)
    plt.ylabel('Political Dimension 2',fontsize=16)
    plt.title(f'Epoch {epochs[idx]} - User coordinates',fontsize=20)
    plt.ylim(0.25,0.75)
    plt.xlim(0.25,0.75)
    plt.xticks([0.5])
    plt.yticks([0.5])
    plt.grid(False)
    plt.show()
    
    

# Ajustar el espacio entre subplots
plt.tight_layout()
plt.show()
```
