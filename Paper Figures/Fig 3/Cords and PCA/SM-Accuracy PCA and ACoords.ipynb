```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# --- Filenames ordered by seed (first 2D, then 1D for each seed) ---
seeds = ['S1', 'S2', 'S3', 'S4', 'S5']
file_names = [(f'{s}Cords.png', f'{s}PCA.png') for s in seeds]

# --- Create 2x5 subplot ---
fig, axes = plt.subplots(2, 5, figsize=(20, 8))

# --- Plot images ---
for col, (file_2d, file_1d) in enumerate(file_names):
    # 2D plot (row 0)
    ax_2d = axes[0, col]
    if os.path.exists(file_2d):
        img_2d = mpimg.imread(file_2d)
        ax_2d.imshow(img_2d)
    else:
        ax_2d.text(0.5, 0.5, f'Missing\n{file_2d}', ha='center', va='center')
    ax_2d.axis('off')
    ax_2d.set_title(f'Seed {col + 1}', fontsize=12)

    # 1D plot (row 1)
    ax_1d = axes[1, col]
    if os.path.exists(file_1d):
        img_1d = mpimg.imread(file_1d)
        ax_1d.imshow(img_1d)
    else:
        ax_1d.text(0.5, 0.5, f'Missing\n{file_1d}', ha='center', va='center')
    ax_1d.axis('off')
    ax_1d.set_title(f'Seed {col + 1}', fontsize=12)

# --- Add row labels to the left side ---
axes[0, 0].text(-0.1, 0.5, '2 Dimension', fontsize=14, rotation=90, va='center', ha='center', transform=axes[0, 0].transAxes)
axes[1, 0].text(-0.1, 0.5, '1 Dimension', fontsize=14, rotation=90, va='center', ha='center', transform=axes[1, 0].transAxes)

output_filename = "combined_seeds_plot.png"
plt.tight_layout()
plt.savefig(output_filename, dpi=300, bbox_inches='tight')
plt.show()

```
